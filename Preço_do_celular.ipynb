```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tqdm import tqdm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/colab/CellPhone_train.csv')
df_valid = pd.read_csv('/content/drive/MyDrive/colab/CellPhone_test.csv')
```

```python
# Dividindo o DataFrame em treinamento (70%) e teste (30%)
train_data, test_data = train_test_split(df, test_size=0.3, random_state=42)
```

```python
train_data.head(2)
```

```python
X_train = train_data.drop(columns=['price_range'])
Y_train = train_data['price_range']

X_test = train_data.drop(columns=['price_range'])
Y_test = train_data['price_range']
```

```python
# Define o nÃºmero de iteraÃ§Ãµes
num_iterations = 100

# Inicializa a melhor precisÃ£o como 0
best_accuracy_RFC = 0
best_RFC = None

# Lista para rastrear a precisÃ£o em cada iteraÃ§Ã£o
accuracy_per_iteration = []

for i in tqdm(range(num_iterations)):
    # Inicializa o classificador RandomForest
    RFC = RandomForestClassifier()

    # Treina o classificador com os dados de treinamento
    RFC.fit(X_train, Y_train)

    # Faz prediÃ§Ãµes nos dados de teste
    y_pred = RFC.predict(X_test)

    # Avalia o desempenho do classificador
    accuracy = accuracy_score(Y_test, y_pred)

    # Adiciona a precisÃ£o atual Ã  lista
    accuracy_per_iteration.append(accuracy)

    # Atualiza a melhor precisÃ£o e o classificador se a precisÃ£o atual for melhor
    if accuracy > best_accuracy_RFC:
        best_accuracy_RFC = accuracy
        best_RFC = RFC

print()
print("Melhor precisÃ£o RFC:", best_accuracy_RFC)

# Encontra a iteraÃ§Ã£o com a melhor precisÃ£o
best_iteration = accuracy_per_iteration.index(best_accuracy_RFC) + 1
print("NÃºmero da melhor iteraÃ§Ã£o:", best_iteration)
```

O modelo alcanÃ§ou uma precisÃ£o de 100% na primeira iteraÃ§Ã£o, isso pode apontar para algumas situaÃ§Ãµes:

1. Overfitting: Talvez o modelo esteja se ajustando demais aos dados de treinamento, o que resulta em uma precisÃ£o perfeita nos dados de teste. Isso pode ocorrer se o modelo for excessivamente complexo em comparaÃ§Ã£o com o tamanho do conjunto de dados.

2. Conjunto de dados pequeno: Se o conjunto de dados for relativamente pequeno, Ã© possÃ­vel que o modelo simplesmente memorize os dados de treinamento, levando a uma precisÃ£o perfeita nos dados de teste.

3. Dados ruidosos: Se os dados contiverem ruÃ­do ou valores discrepantes (outliers), o modelo pode se ajustar a esses padrÃµes de ruÃ­do e, consequentemente, obter uma alta precisÃ£o nos dados de teste.

4. Possibilidade de vazamento de dados: Ã‰ crucial garantir que nÃ£o haja vazamento de dados entre os conjuntos de treinamento e teste. Se houver vazamento de dados, o modelo pode parecer ter uma precisÃ£o perfeita, mas falharÃ¡ em generalizar bem para novos dados.

Ã‰ essencial avaliar cuidadosamente o desempenho do modelo e considerar esses fatores ao interpretar os resultados. Se uma precisÃ£o de 100% parecer improvÃ¡vel ou nÃ£o representativa do verdadeiro desempenho do modelo, Ã© aconselhÃ¡vel investigar mais a fundo.

```python
from sklearn.neighbors import KNeighborsClassifier

# Define the number of iterations
num_iterations = 100

best_accuracy = 0
best_knn = None

for i in tqdm(range(num_iterations)):
    # Initialize the KNN classifier
    knn = KNeighborsClassifier(n_neighbors=8)

    # Train the classifier on the training data
    knn.fit(X_train, Y_train)

    # Make predictions on the testing data
    y_pred = knn.predict(X_test)

    # Evaluate the classifier's performance
    accuracy = accuracy_score(Y_test, y_pred)
    #print(f"Iteration {i+1}: Accuracy test - {accuracy}")

    # Update best accuracy and classifier if current accuracy is better
    if accuracy > best_accuracy:
        best_accuracy_knn = accuracy
        best_knn = knn

print()
print("Best accuracy knn:", best_accuracy_knn)
```
