```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tqdm import tqdm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/drive/MyDrive/colab/CellPhone_train.csv')
df_valid = pd.read_csv('/content/drive/MyDrive/colab/CellPhone_test.csv')
```

```python
# Dividindo o DataFrame em treinamento (70%) e teste (30%)
train_data, test_data = train_test_split(df, test_size=0.3, random_state=42)
```

```python
train_data.head(2)
```

```python
X_train = train_data.drop(columns=['price_range'])
Y_train = train_data['price_range']

X_test = train_data.drop(columns=['price_range'])
Y_test = train_data['price_range']
```

```python
# Define o número de iterações
num_iterations = 100

# Inicializa a melhor precisão como 0
best_accuracy_RFC = 0
best_RFC = None

# Lista para rastrear a precisão em cada iteração
accuracy_per_iteration = []

for i in tqdm(range(num_iterations)):
    # Inicializa o classificador RandomForest
    RFC = RandomForestClassifier()

    # Treina o classificador com os dados de treinamento
    RFC.fit(X_train, Y_train)

    # Faz predições nos dados de teste
    y_pred = RFC.predict(X_test)

    # Avalia o desempenho do classificador
    accuracy = accuracy_score(Y_test, y_pred)

    # Adiciona a precisão atual à lista
    accuracy_per_iteration.append(accuracy)

    # Atualiza a melhor precisão e o classificador se a precisão atual for melhor
    if accuracy > best_accuracy_RFC:
        best_accuracy_RFC = accuracy
        best_RFC = RFC

print()
print("Melhor precisão RFC:", best_accuracy_RFC)

# Encontra a iteração com a melhor precisão
best_iteration = accuracy_per_iteration.index(best_accuracy_RFC) + 1
print("Número da melhor iteração:", best_iteration)
```

O modelo alcançou uma precisão de 100% na primeira iteração, isso pode apontar para algumas situações:

1. Overfitting: Talvez o modelo esteja se ajustando demais aos dados de treinamento, o que resulta em uma precisão perfeita nos dados de teste. Isso pode ocorrer se o modelo for excessivamente complexo em comparação com o tamanho do conjunto de dados.

2. Conjunto de dados pequeno: Se o conjunto de dados for relativamente pequeno, é possível que o modelo simplesmente memorize os dados de treinamento, levando a uma precisão perfeita nos dados de teste.

3. Dados ruidosos: Se os dados contiverem ruído ou valores discrepantes (outliers), o modelo pode se ajustar a esses padrões de ruído e, consequentemente, obter uma alta precisão nos dados de teste.

4. Possibilidade de vazamento de dados: É crucial garantir que não haja vazamento de dados entre os conjuntos de treinamento e teste. Se houver vazamento de dados, o modelo pode parecer ter uma precisão perfeita, mas falhará em generalizar bem para novos dados.

É essencial avaliar cuidadosamente o desempenho do modelo e considerar esses fatores ao interpretar os resultados. Se uma precisão de 100% parecer improvável ou não representativa do verdadeiro desempenho do modelo, é aconselhável investigar mais a fundo.

```python
from sklearn.neighbors import KNeighborsClassifier

# Define the number of iterations
num_iterations = 100

best_accuracy = 0
best_knn = None

for i in tqdm(range(num_iterations)):
    # Initialize the KNN classifier
    knn = KNeighborsClassifier(n_neighbors=8)

    # Train the classifier on the training data
    knn.fit(X_train, Y_train)

    # Make predictions on the testing data
    y_pred = knn.predict(X_test)

    # Evaluate the classifier's performance
    accuracy = accuracy_score(Y_test, y_pred)
    #print(f"Iteration {i+1}: Accuracy test - {accuracy}")

    # Update best accuracy and classifier if current accuracy is better
    if accuracy > best_accuracy:
        best_accuracy_knn = accuracy
        best_knn = knn

print()
print("Best accuracy knn:", best_accuracy_knn)
```
